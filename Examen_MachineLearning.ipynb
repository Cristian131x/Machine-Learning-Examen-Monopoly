#### **Regresi√≥n Logistica**üì¶

Importamos las librer√≠as necesarias:

```python
from sklearn import model_selection
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

Establecemos las variables necesarias para implementar en nuestro modelo y como ir√° distribuida nuestros test y train.

```python
dfNuevo = dfF[dfF['Cuentas']<3]
```

```python
X = dfNuevo[['CUPO_L1','CUPO_MX']]
y = dfNuevo['Cuentas']

X_train, X_test, Y_train, Y_test = model_selection.train_test_split(X, y, test_size=0.3,random_state=7)
```

Ac√° se realiza la creaci√≥n del modelo junto al entrenamiento de los datos con base en la regresi√≥n log√≠stica que es un modelo de clasificaci√≥n se destaca que se realiza altas interacciones por errores.

```python
# Creamos nuestro modelo
std = StandardScaler()

# Realizamos el entrenamiento de los datos
X_train = std.fit_transform(X_train)
X_validation = std.transform(X_test)

# Se destaca que las interacciones deben ser altas si no aparecen diferentes errores
model = linear_model.LogisticRegression(max_iter=50000)
model.fit(X_train, Y_train)
```

#####¬†**M√©tricas¬†de¬†Rendimiento¬†‚è≥**
Ac√°¬†tenemos¬†nuestras¬†m√©tricas¬†de¬†rendimiento¬†que¬†nos¬†ayudaran¬†a¬†saber¬†si¬†nuestro¬†modelo¬†de¬†**Regresi√≥n¬†Log√≠stica**¬†es¬†el¬†correcto.

```python
name='Regresi√≥n Logistica'
kfold = model_selection.KFold(n_splits=10, random_state=7, shuffle=True)
cv_results = model_selection.cross_val_score(model, X_train, Y_train, cv=kfold, scoring='accuracy')
msg = "%s: %f (%f)" % (name, cv_results.mean(), cv_results.std())
print(msg)

print('Puntaje del Modelo: %.2f' % model.score(X_train, Y_train))

predictions = model.predict(X_validation)
training_prediction = model.predict(X_train)

print('Puntaje de Precisi√≥n Tester: %.2f' % accuracy_score(Y_test, predictions))
print('Puntaje de Precisi√≥n Training: %.2f' % accuracy_score(Y_train, training_prediction))
print('')
print('Matriz de confusi√≥n:')
print(confusion_matrix(Y_train, training_prediction))
print('')
print('Reporte de clasificaci√≥n:')
print(classification_report(Y_test, predictions))
```

##### **CrossValidation** üê±‚ÄçüöÄ
